```python
import matplotlib.pyplot as plt
import numpy as np
import scipy

x = [1,2,2.5,3,4]
y = [1,4,7,9,15]
plt.plot(x, y, 'yo')
plt.axis([0,6,0,20])         # y = mx + b    m;Slope, y&x;Axis b:Intercept
```

### Line of Regression

```python
plt.plot(x, y, 'ro')
plt.axis([0,6,0,20])
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))
plt.show()
```
